```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\SRIKANT SENAPATI\OneDrive\Desktop\Sample_Superstore.csv",encoding='latin1')
df.head(2)
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df['order_date'] = df['order_date'].str.replace('-','/')
df['ship_date'] = df['ship_date'].str.replace('-','/')
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')
```

```python
df.columns
```

```python
df = df.drop(columns=['postal_code', 'country'])
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
df.to_csv(r"C:\Users\SRIKANT SENAPATI\OneDrive\Desktop\cleaned_Superstore.csv", index=False)
```
